```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import math
 
#define aim function
def aimFunction(x):
#     if (x > 80):
    y =  x**3-60*x**2-4*x+6 - math.sin(x/5)*100000
#     else:
#         y=x**3-60*x**2-4*x+6
    return y
x=[i/10 for i in range(1000)]
y=[0 for i in range(1000)]
for i in range(1000):
    y[i]=aimFunction(x[i])
 
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
T=1000 #initiate temperature
Tmin=10 #minimum value of terperature
x=np.random.uniform(low=0,high=100)#initiate x
k=50 #times of internal circulation 
y=0#initiate result
t=0#time

x_list = list()
while T>=Tmin:
    for i in range(k):
        #calculate y
        y=aimFunction(x)
        #generate a new x in the neighboorhood of x by transform function
        xNew=x+np.random.uniform(low=-0.055,high=0.055)*T
        if (0<=xNew and xNew<=100):
            yNew=aimFunction(xNew)
            if yNew-y<0:
                x=xNew
            else:
                #metropolis principle
                p=math.exp(-(yNew-y)/T)
                r=np.random.uniform(low=0,high=1)
                if r<p:
                    x=xNew
                    
        x_list.append(x)
    t+=1
#     print (t)
    T=1000/(1+t)
    
print (x,aimFunction(x))
```

```python
negative_x_list = list()
for x_value in x_list:
    negative_x_list.append(x_value * -1)
```

```python
step_index = list(range(len(x_list)))
negative_step_index = list()
for index in step_index:
    negative_step_index.append(index * -1)
```

```python
fig = plt.figure(figsize=(10,5))
plt.xlim(0,100)
plt.plot(x_list[0:500], negative_step_index[0:500] )
plt.show()
```
